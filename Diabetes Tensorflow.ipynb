```python
import pandas as pd
my_name = ["Pregnancies","Glucose","BloodPressure",
           "SkinThickness","Insulin","BMI","DiabetesPedigreeFunction",
           "Age","Class"]

df = pd.read_csv("pima-indians-diabetes.csv",header=0,names=my_name)
```

```python
print(df.columns)
```

```python
col_norm =['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction']
```

```python
df1 = df[col_norm].apply(lambda x : (( x - x.min() ) / ( x.max() - x.min() )))
print(df.head(2))
```

```python
df1.columns
```

```python
df1.head(3)
```

```python
import tensorflow as tf
```

```python
num_preg = tf.feature_column.numeric_column('Pregnancies')
plasma_gluc = tf.feature_column.numeric_column('Glucose')
dias_press = tf.feature_column.numeric_column('BloodPressure')
num_tricep = tf.feature_column.numeric_column('SkinThickness')
num_insulin = tf.feature_column.numeric_column('Insulin')
num_bmi = tf.feature_column.numeric_column('BMI')
num_peggree  = tf.feature_column.numeric_column('DiabetesPedigreeFunction')
```

```python
assigned_group = tf.feature_column.categorical_column_with_vocabulary_list('Class',['A','B'])
```

```python
feat_col = [num_preg,plasma_gluc,dias_press,num_tricep,num_insulin,num_bmi,num_peggree]
```

```python
x_data = df1
x_data.head(3)
```

```python
label = df["Class"]
print(label.head(4))
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_data,label, test_size=0.3,random_state=101)
```

```python
 input_func = tf.estimator.inputs.pandas_input_fn(x=x_train,y=y_train,batch_size=10,num_epochs=1000,shuffle=True)
```

```python
model = tf.estimator.LinearClassifier(feature_columns=feat_col,n_classes=2)
```

```python
model.train(input_fn=input_func,steps=1000)
```

```python
eval_input_func = tf.estimator.inputs.pandas_input_fn(x=x_test,y=y_test,batch_size=10,num_epochs=1,shuffle=False)
```

```python
results = model.evaluate(eval_input_func)
```

```python
results
```

```python
pred_input_func = tf.estimator.inputs.pandas_input_fn(x=x_test,batch_size=10,num_epochs=1,shuffle=False)
```

```python
rediction = model.predict(pred_input_func)
```

```python
neural_predicted = list(rediction)
```
